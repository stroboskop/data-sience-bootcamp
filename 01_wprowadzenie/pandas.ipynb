```python
import pandas as pd
import numpy as np


#series
s = pd.Series(data=[3, 2, 4, 6])
s

s = pd.Series(data=[3, 2, 4, 6], index=['a', 'b', 'c', 'd'], name='sample')
s

#brak danych
np.nan

s = pd.Series(data=[3, np.nan, 4, 6], index=['a', 'b', 'c', 'd'], name='sample')
s

s = pd.Series(data=[True, False, False])
s

s = pd.Series(data=np.arange(10,20), index=pd.date_range(start='20211207',periods=10))
s

#wyświetli nam liste obiektów
s.index

#wyświetli nam listę z wartościami
s.values

#wyświetli nam typ 'o' - to strin-> Object
s.dtypes

#kształ danych
s.shape


#praca z obiektami
price = pd.Series(data={'Apple': 200, 'CD Project': 60, 'Amazon': 1900, 'KGHM': np.nan})
price

price[1]

#PODSTAWOWE FUNCKCJE ANLIZY DANYCH

#zliczanie
price.count()

#zwaraca ile razy wystapiła dana wartość
price.value_counts()

price.min()
price.max()
price.sum()

#wyświetla statystki zbiorcze
price.describe()

#wyświetl n największych elementów
price.nlargest(1)

#wyświetl n najmniejeszych elementów
price.nsmallest(1)

#zwróci posortowane elementy
price.rank()

#zwróci posortowane wartości
price.sort_values(ascending=False)

#METODA APPLY
#za pomocą tej metody możemy zmieniać wartości elementów, np w tym przypadku
#zwiększyć wartość 3,8 raza
price = price.apply(lambda x: x * 3.8)
price


#DATAFRAME
df = pd.DataFrame(data=[12, 12, 32], index=['first', 'second', 'third'], columns=['col_1'])
df


#w tym przypadku nazwy kolumn są automatycznie określone przez nazwy kluczy
df = pd.DataFrame(data={'WIG20': ['PKN ORLEN', 'PKO BP'],
                        'mWIG40': ['Amica', 'Playway']})
df

df = pd.DataFrame(data=[[10, 12, 13], [23, 12, 10]], index=['first', 'second'], columns=['col_1', 'col_2', 'col_3'])
df

# df.columns
# df.index
# df.values
# df.info()
#podaje dataFrame ze statystykami, można ją transponować
# df.describe()


#SELEKCJA KOLUMN
#zwróci nam int64
df['col_1']
type(df['col_1'])

#zwróci nam dataFrame
df[['col_1']]
type(df[['col_1']])

#zmiana nazwy kolumn
df.columns =  ['a', 'b', 'c']
df

#możemy odwołać się bezpośrednio bo nazwy kolumny, a co gdy zmieni nam się nazwa kolumny?
df.a

#stworzyliśmy właśnie nową kolumne
df['d'] = df.a + df.c
df


#metoda lock
df = pd.DataFrame(data=[[10, 12, 13], [23, 12, 10]], index=['first', 'second'], columns=['col_1', 'col_2', 'col_3'])
df.loc['first']
df.iloc[0]

#zwróci 12
df.loc['first', 'col_2']

#zwróci 12
df.loc[:, 'col_2']
```
