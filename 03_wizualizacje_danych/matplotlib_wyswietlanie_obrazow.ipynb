```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

```

# Wyświetlanie obrazów i subploty

```python
img = mpimg.imread('office.jpg')
img
```

```python
plt.imshow(img  )
```

```python
fig = plt.figure(figsize=(12,10))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
  plt.subplot(140 + idx + 1)
  plt.imshow(mpimg.imread(img))
```

```python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)


plt.style.use('ggplot')
plt.plot(x, x_sin)

plt.style.available
```
